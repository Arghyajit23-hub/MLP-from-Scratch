```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('reg_dataset.csv')
```

```python
df
```

```python
def loss(y,y_hat):
    return (y-y_hat)**2
```

```python
def MLP_Reg(lr,epoch,tol,df):
    
    ## Initialization
    w_1_11=0
    w_1_12=0
    w_1_21=0
    w_1_22=0
    w_2_11=0
    w_2_21=0
    b_11=1
    b_12=1
    b_21=1
    
    l=len(df)
    
    n=lr  ## assigning learning rate
    

    ## First epoch to evaluate first loss
    avg_loss=[]
    for i in range(l):
        o_11=w_1_11*df.iloc[i,0] + w_1_21*df.iloc[i,1] + b_11
        o_12=w_1_12*df.iloc[i,0] + w_1_22*df.iloc[i,1] + b_12
        o_21=w_2_11*o_11 + w_2_21*o_12 + b_21
            
        y=df.iloc[i,2]
        y_hat=o_21
            
        mse=loss(y,y_hat)
        avg_loss.append(mse)
            
        w_1_11=w_1_11 - n*(-2*(y-y_hat)*w_2_11*df.iloc[i,0])
        w_1_21=w_1_21 - n*(-2*(y-y_hat)*w_2_11*df.iloc[i,1])
        b_11=b_11 - n*-2*(y-y_hat)*w_2_11
            
        w_1_12=w_1_12 - n*(-2*(y-y_hat)*w_2_21*df.iloc[i,0])
        w_1_22=w_1_22 - n*(-2*(y-y_hat)*w_2_21*df.iloc[i,1])
        b_12=b_12 - n*(-2)*(y-y_hat)*w_2_21
            
        w_2_11=w_2_11 - n*(-2)*o_11*(y-y_hat)
        w_2_21=w_1_21 - n*(-2)*o_12*(y-y_hat)
        b_21=b_21 - n*(-2)*(y-y_hat)
    
    prev_loss=np.mean(avg_loss)
    print(f"Epoch {1} --> Loss:{prev_loss}")
    
    for j in range(1,epoch):   ## starts from 1 instead of 0 since first epoch has been calculated previously
        avg_loss=[]
        for i in range(l):
            o_11=w_1_11*df.iloc[i,0] + w_1_21*df.iloc[i,1] + b_11
            o_12=w_1_12*df.iloc[i,0] + w_1_22*df.iloc[i,1] + b_12
            o_21=w_2_11*o_11 + w_2_21*o_12 + b_21
            
            y=df.iloc[i,2]
            y_hat=o_21
            
            mse=loss(y,y_hat)
            avg_loss.append(mse)
            
            w_1_11=w_1_11 - n*(-2*(y-y_hat)*w_2_11*df.iloc[i,0])
            w_1_21=w_1_21 - n*(-2*(y-y_hat)*w_2_11*df.iloc[i,1])
            b_11=b_11 - n*-2*(y-y_hat)*w_2_11
            
            w_1_12=w_1_12 - n*(-2*(y-y_hat)*w_2_21*df.iloc[i,0])
            w_1_22=w_1_22 - n*(-2*(y-y_hat)*w_2_21*df.iloc[i,1])
            b_12=b_12 - n*(-2)*(y-y_hat)*w_2_21
            
            w_2_11=w_2_11 - n*(-2)*o_11*(y-y_hat)
            w_2_21=w_1_21 - n*(-2)*o_12*(y-y_hat)
            b_21=b_21 - n*(-2)*(y-y_hat)
    
        new_loss=np.mean(avg_loss)  
        print(f"Epoch {j+1} --> Loss:{new_loss}")   ## epoch counting starts from 2
        
        if abs(new_loss-prev_loss)<tol:
            print(new_loss)
            break

        prev_loss=new_loss
        
    my_dict={'w_1_11':w_1_11, 'w_1_12':w_1_12, 'w_1_21':w_1_21, 'w_1_22':w_1_22, 'w_2_11':w_2_11, 'w_2_21':w_2_21, 'b_11':b_11, 'b_12':b_12, 'b_21':b_21}

    return my_dict
```

```python
lr=0.0001
epoch=100
tol=0.001

MLP_Reg(lr,epoch,tol,df)
```
